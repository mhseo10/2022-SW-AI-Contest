```python
label_dir = './[라벨]Training/1.간판/1.가로형간판/가로형간판1\\'
train_dir = './[원천]Training_간판_가로형간판_원천데이터1/'
def crop_label(bbox):
    bbox[2]+= bbox[0]
    bbox[3]+= bbox[1]
    return bbox
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
x = 0 #확인하고 싶은 index

with open(label_dir+os.listdir(label_dir)[x],encoding='utf8') as f:
    j = json.load(f)
image = Image.open(f'{train_dir+os.listdir(train_dir)[x]}')
crop= []
text = []
for idx,i in enumerate(j['annotations']):
    crop.append(i['bbox'])
    text.append(i['text'])
    crop[idx] = crop_label(crop[idx])
f,ax = plt.subplots(1,len(crop),figsize = (30,10))

if len(crop)>1:
    for k in range(len(crop)):
        ax[k].set_title(f'{text[k]}',fontsize='50')
        ax[k].axis(False)
        ax[k].imshow(image.crop(crop[k]))
        f.tight_layout
else: 
    ax.set_title(f'{text[idx]}',fontsize='50')
    ax.imshow(image.crop(crop[idx]))
    ax.axis(False)
    # crop = [9,636,102,78]


```
